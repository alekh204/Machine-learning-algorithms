```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= (20.0, 10.0)
             
#Reading data
data = pd.read_csv('weatherHistory.csv')
print(data.shape)
data.head()
```

```python
X= data['Humidity'].values
Y= data['Temperature (C)'].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)


n = len(X)

slope= 0
slope_lower= 0
for i in range(n):
    slope +=(X[i] - mean_x)* (Y[i] - mean_y)
    slope_lower += (X[i] - mean_x) ** 2
b1=slope/slope_lower
b0=mean_y - (b1 * mean_x)
print(b1,b0)
```
